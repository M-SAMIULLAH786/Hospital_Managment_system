## 7. Bus Stops

I made a list for stops. Got fastest, time, and stops.

```python
stops = [
    {"name": "A", "time": 0, "dist": 0},
    {"name": "B", "time": 15, "dist": 5},
    {"name": "C", "time": 30, "dist": 12},
    {"name": "D", "time": 50, "dist": 20}
]

times = []
for i in range(len(stops) - 1):
    t = stops[i + 1]["time"] - stops[i]["time"]
    times.append(t)

fastest = min(times)
total_time = stops[-1]["time"] - stops[0]["time"]
total_stops = len(stops)

print("Fastest", fastest)
print("Total Time:", total_time)
print("Total Stop:", total_stops)
```

## 9. Tournament

I used a list for scores. Got winner and average.

```python
games = [
    {"team1": "A", "team2": "B", "score1": 3, "score2": 2},
    {"team1": "B", "team2": "C", "score1": 1, "score2": 2},
    {"team1": "A", "team2": "C", "score1": 2, "score2": 2}
]

points = {}
for g in games:
    t1 = g["team1"]
    t2 = g["team2"]
    s1 = g["score1"]
    s2 = g["score2"]

    if t1 in points:
        points[t1] += s1
    else:
        points[t1] = s1

    if t2 in points:
        points[t2] += s2
    else:
        points[t2] = s2

winner = max(points, key=points.get)
avg = sum(points.values()) / len(points)
teams = set(points.keys())

print("Winner:", winner)
print("Avg:", avg)
print("Teams:", teams)
```

## 11. Attendance

I used a dictionary. Found full, most absent, and percent.

```python
att = {
    "sami": [1]*30,
    "aneeq": [1]*28 + [0,0],
    "azan": [1]*25 + [0]*5
}

full = []
for name, days in att.items():
    if sum(days) == 30:
        full.append(name)

absent = {}
for name, days in att.items():
    absent[name] = days.count(0)

most = max(absent, key=absent.get)

percent = {}
for name, days in att.items():
    percent[name] = (sum(days) / 30) * 100

print("Full:", full)
print("Most Absent:", most)
print("Percent:", percent)
```

## 14. Hotel

I used a dictionary. Checked free, busy, and cost.

```python
rooms = {
    101: {"busy": True, "rate": 3000, "days": 2},
    102: {"busy": False, "rate": 2000, "days": 0},
    103: {"busy": True, "rate": 2500, "days": 3}
}

free = []
busy = []
cost = {}

for r, d in rooms.items():
    if d["busy"]:
        busy.append(r)
        cost[r] = d["rate"] * d["days"]
    else:
        free.append(r)

print("Free:", free)
print("Busy:", busy)
print("Cost:", cost)
```

## 15. Expenses

I used a list. Got total, checked budget, found largest.

```python
spend = [
    {"cat": "food", "amt": 500},
    {"cat": "travel", "amt": 1200},
    {"cat": "food", "amt": 800},
    {"cat": "fun", "amt": 600}
]

budget = 2500
total = {}

for s in spend:
    cat = s["cat"]
    amt = s["amt"]
    if cat in total:
        total[cat] += amt
    else:
        total[cat] = amt

sum_amt = 0
for s in spend:
    sum_amt += s["amt"]

over = sum_amt > budget

big_amt = 0
big_cat = ""
for s in spend:
    if s["amt"] > big_amt:
        big_amt = s["amt"]
        big_cat = s["cat"]

print("Total:", total)
print("Over Budget:", over)
print("Biggest:", {"cat": big_cat, "amt": big_amt})
```

## 16. Business

I made functions to add and show money. Got totals and biggest.

```python
trans = []

def add(date, kind, amt, cat):
    trans.append({
        "date": date,
        "kind": kind,
        "amt": amt,
        "cat": cat
    })

def sums():
    total = {}
    for t in trans:
        cat = t["cat"]
        amt = t["amt"]
        if cat in total:
            total[cat] += amt
        else:
            total[cat] = amt
    return total

def big():
    top = trans[0]
    for t in trans:
        if t["amt"] > top["amt"]:
            top = t
    return top

def show():
    total = sums()
    biggest = big()
    print("Totals:", total)
    print("Biggest:", biggest)

add("2025-01-01", "out", 500, "food")
add("2025-01-02", "in", 2000, "salary")
add("2025-01-03", "out", 700, "travel")

show()
```
